## Exercise 1
Use Numpy to create two sets of 1000 Datapoints. Both should be following a normal distribution, one around an expected value of 5 with a standard deviation of 1.5, the other around an expected value of 12 with a standard deviation of 3.  

Plot both sets in one histogram.

```python
#Add code to create both datasets
dataset1 = np.random.normal(5, 1.5, 1000)
dataset2 = np.random.normal(12, 3, 1000)

#Add code to plot both in a histogram
fig = go.Figure()
fig.add_trace(go.Histogram(x=dataset1, name="mu = 5, sigma = 1.5"))
fig.add_trace(go.Histogram(x=dataset2, name="mu = 12, sigma = 3"))

# Overlay both histograms
fig.update_layout(barmode='overlay',
                title="Different Normal Distributions",
                xaxis_title="x",
                yaxis_title="count",
                font=dict(
                        size=18,
                        color="navy")
                    )
# Reduce opacity to see both histograms
fig.update_traces(opacity=0.75)
fig.show()
```

Create a third set of Datapoints containing the points of both of the previous sets. Plot this Dataset as a histogram as well.

```python
#Add code to combine the two existing datasets
dataset3 = np.concatenate((dataset1, dataset2), axis = 0)
#Add code to plot the result as a histogramm
fig2 = go.Figure()
fig2.add_trace(go.Histogram(x=dataset3, name="combined dataset"))
fig2.update_layout(
                title="Combined Distribution",
                xaxis_title="x",
                yaxis_title="count",
                font=dict(
                        size=18,
                        color="navy")
                    )
fig2.show()
```

You should see a histogram with two "humps". However the second hump should be significantly smaller than it was when plotting the datasets separately. Why is that the case?

-> Both datasets contain equally 1000 observations, however for the second set with a greater sigma, the data is spread over a wider range. Consequently, there are more observations in the surrounding of the expected value of the first dataset. So since with the same bin size, the bins closer to set one contain more abservations.

## Exercise 2
Create a function that calculates the mean and standard deviation of datasets and uses those values to calculate the `SelfExp` for each value of the given dataset. Use this function to calculate the `SelfExp` for each value of each of the three datasets.  
Next create a scatterplot with the selfexp value as the y-axis, and the actual randomly generated value on the x-axis

```python
def selfExp(values):
    """DOCSTRING:
    calculate standardized residuals of input numpy array values
    """
    # mean
    mean = np.mean(values)

    # standard deviation
    sigma = np.std(values)
    #Add code for calculating the selfexp values
    s = np.absolute(values - mean)/(sigma)

    # threshold values only highlight 1% of the cells

    return s
```

```python
#Add code to create a scatterplot

i = 1
for x in [dataset1, dataset2, dataset3]:

    s = selfExp(x)

    fig = go.Figure(data=go.Scatter(x=x, y=s, mode='markers'))
    fig.update_layout(
                    title="Standardized Residuals dataset "+str(i),
                    xaxis_title="x",
                    yaxis_title="residuals s",
                    font=dict(
                            size=18,
                            color="navy")
                        )

    fig.show()
    i+=1
```

The resulting plot does not necessarily grant us new insight into the properties of our dataset. It should however give us a nice visualization and hopefully some understanding of what the selfexp actually is.

Now consider a threshold value $\tau = 2.5$, representing a probability of $99\%$ (compare slide 32 of the lecture).
For each of our three datasets, create a new dataset only containing the values with a SelfExp $s>\tau$.

```python
i = 1
for x in [dataset1, dataset2, dataset3]:

    s = selfExp(x)
    #Add code to filter the datasets

    # get only values where threshold condition is fullfilled
    filtered_s = s[s>2.5]
    filtered_x = x[s>2.5]

    fig = go.Figure(data=go.Scatter(x=filtered_x, y=filtered_s, mode='markers'))
    fig.update_layout(
                    title="Standardized Residuals dataset "+str(i),
                    xaxis_title="x",
                    yaxis_title="residuals s",
                    font=dict(
                            size=18,
                            color="navy")
                        )
    fig.show()
    i+=1
```
