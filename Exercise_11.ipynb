```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## ROC Curve

```python
scores = np.array([0.9,0.8,0.7,0.6,0.5,0.45,0.4,0.3,0.2,0.1])
true_labels = np.array([1,1,0,1,0,0,1,0,1,0])


#def classifier_K_tau(scores):

fig = plt.figure()

TPR = np.array([])
FPR = np.array([])

tau = scores
N = np.size(tau)
for t in tau:
    # calculate predeiction according to given threshold
    prediction = scores >= t

    # true positive rate, 
    tpr = ((prediction == true_labels)&(prediction == True)).sum()/N
    TPR = np.append(TPR, tpr)

    # false positive rate
    fpr = ((prediction != true_labels)&(prediction == True)).sum()/N
    FPR = np.append(FPR, fpr)

    plt.scatter(t, tpr, c='b')
    plt.scatter(t, fpr, c='r')

plt.legend(['true positive rate', 'false positive rate'])
plt.xlabel('tau')
plt.show()
    

```

```python
ROC_plot = plt.figure(figsize=[16,9])

sns.scatterplot(FPR,TPR)
for i in range(len(scores)):
    plt.text(FPR[i],TPR[i],' t = '+str(scores[i]))
# bisecting line
bisecting = np.linspace(0,1,10)
# add the label for tau
sns.lineplot(bisecting,bisecting)

plt.xlim(0,1)
plt.ylim(0,1)
plt.title('ROC space')
plt.xlabel('fpr')
plt.ylabel('tpr')
plt.show()
```

Best choice is probably tau = 0.6.

```python
fig = plt.figure()

for t in tau:
    # calculate predeiction according to given threshold
    prediction = scores >= t

    # accurracy, 
    acc = (prediction == true_labels).sum()/N

    # error rate
    err = (prediction != true_labels).sum()/N

    plt.scatter(t, acc, c='b')
    plt.scatter(t, err, c='r')

plt.legend(['accuracy', 'error rate'])
plt.xlabel('tau')
plt.show()
```
